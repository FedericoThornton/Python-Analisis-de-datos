```python
#Regresion
import pandas as pd
df = pd.read_csv('/content/Copia_de_boston_final.csv')
```

```python
# variables predictoras x, variable target y

X = df.drop('MEDV', axis=1) #sin inplace=true
y = df['MEDV']
```

```python
y
```

```python
#realizar train test split con porcentuales indicados (75-25)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=11)
```

```python
#tama√±o de cada subset
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

```python
x_train.head()
```

```python
#regresion lineal
from sklearn.linear_model import LinearRegression
#regresion lineal sin parametros
regresion = LinearRegression()
```

```python
regresion
```

```python
# entrenar la regresion lineal con el set de entrenamiento metodo fit entrenar
regresion.fit(x_train, y_train)
```

```python
x_test['Y_pred'] = regresion.predict(x_test)
```

```python
x_test.head()
```

```python
#realizar una prediccion
regresion.predict(x_test)
```

```python
#evaluacion MAE R2
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
#error medio absoluto
x_test['y_real'] =y_test
x_test.head(10)
```

```python
#error abs es el valor absoluto
x_test['error'] = abs(x_test['y_real'] - x_test['Y_pred'])
x_test.head(10)
```

```python
x_test.error.mean()
```

```python
print("Yreal minimo", str(x_test.y_real.min()))
print("Yreal maximo", str(x_test.y_real.max()))
```

```python
#Realizar evaluacion MAE
mean_absolute_error(x_test.y_real, x_test.Y_pred)
```

```python
#realizar evaluacion R2
r2_score(x_test.y_real, x_test.Y_pred)
```
