```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
le_Embarked = LabelEncoder()
```

```python
df['sex_encoded'] = le_sex.fit_transform(df['Sex'])
df['Embarked_encoded'] = le_Embarked.fit_transform(df['Embarked'])
```

```python
lista_borrado = ['PassengerId', 'Name', 'Cabin', 'Ticket','Sex', 'Embarked']
df.drop(lista_borrado, axis=1, inplace=True)
```

```python
X = df.drop('Survived', axis=1)
df.head()
```

```python
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=11)
print("Shape de X Train ", X_train.shape)
print("Shape de X test ", X_test.shape)
print("Shape de y train ", y_train.shape)
print("Shape de y test ",  y_test.shape)
```

```python
from sklearn import tree
arbol = tree.DecisionTreeClassifier()
```

```python
arbol.fit(X_train, y_train)
```

```python
X_test.drop('Ticket', axis=1, inplace=True)
```

```python
X_test["y_pred"] = arbol.predict(X_test)
```

```python
X_test['y_real'] = y_test
X_test.head(10)
```

```python
output = X_test[['y_pred', 'y_real']]
```

```python
output
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
#visualizador
import graphviz
```

```python
#accuracy aciertos
accuracy_score(X_test['y_pred'], X_test['y_real'])
```

```python
cm = confusion_matrix(X_test['y_pred'], X_test['y_real'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=arbol.classes_)
disp.plot()

plt.show()
```

```python
#matriz de confusion falso negativo cuando dice que no sobrevive y si sobrevive y el falso positivo porque dice que sobrevive y en realidad no.
from sklearn.ensemble import RandomForestClassifier
```

```python
#eliminar y pred e y real porque las toma como variables nuevamente.
X_test_limpio = X_test.drop(['y_pred', 'y_real'], axis=1)
```

```python
#bagging
clf = RandomForestClassifier(n_estimators=300)
clf.fit(X_train, y_train)
y_pred_bagging = clf.predict(X_test_limpio)
```

```python
#boosting
!pip install xgboost
from xgboost import XGBClassifier
xgboost = XGBClassifier(n_estimators = 300)
xgboost.fit(X_train, y_train)
y_pred_boosting = xgboost.predict(X_test_limpio)
```

```python
print("Accuracy Bagging: " + str(accuracy_score(y_pred_bagging, y_test)))
print("Accuracy Boosting: " + str(accuracy_score(y_pred_boosting, y_test)))
```
